# Telecom Churn - ML Group Case Study

## Problem Statement
### In the telecom industry, customers are able to choose from multiple service providers and actively switch from one operator to another. In this highly competitive market, the telecommunications industry experiences an average of 15-25% annual churn rate. Given the fact that it costs 5-10 times more to acquire a new customer than to retain an existing one, customer retention has now become even more important than customer acquisition. 

### To reduce customer churn, telecom companies need to predict which customers are at high risk of churn.

## Business Goal 

### In this project, we will analyse customer-level data of a leading telecom firm, build predictive models to identify customers at high risk of churn and identify the main indicators of churn.

```python
#Calculate percentage of null values in all columns
def missing_values(df):
    null_values = df.isnull().sum(axis=0).sort_values(ascending=False)
    percent_null_values = (df.isnull().sum(axis=0)/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([null_values, percent_null_values], axis=1, keys=['Total', 'Percent'])
    
    return missing_data
```

### Define derived features

```python
#Minutes per day - Total out going calls / age on network
churn_data['minutes_per_day'] = (churn_data['total_og_mou_6']+churn_data['total_og_mou_7']+
                                 churn_data['total_og_mou_8'])/churn_data['aon']
```

```python
#Minutes to internet - Total out going calls / total mobile internet usage volume on 2G and 3G networks 
churn_data['minutes_to_internet'] = ((churn_data['total_og_mou_6']+churn_data['total_og_mou_7']+churn_data['total_og_mou_8'])
/
(churn_data['vol_2g_mb_6']+churn_data['vol_2g_mb_7']+churn_data['vol_2g_mb_8']+
 churn_data['vol_3g_mb_6']+churn_data['vol_3g_mb_7']+churn_data['vol_3g_mb_8']+1))
```

```python
#Off net call ratio - If offnet call calls are increasing then the user is talking more with the customers on other networks.
#That means that user's friends have already switched to other networks and there may be a social influence on the user
#to churn
churn_data['offnet_call_ratio_6'] = churn_data['offnet_mou_6']/(churn_data[['onnet_mou_6','offnet_mou_6']].sum(axis=1) + 1)
churn_data['offnet_call_ratio_7'] = churn_data['offnet_mou_7']/(churn_data[['onnet_mou_7','offnet_mou_7']].sum(axis=1) + 1)
churn_data['offnet_call_ratio_8'] = churn_data['offnet_mou_8']/(churn_data[['onnet_mou_8','offnet_mou_8']].sum(axis=1) + 1)
```

```python
# Delta from Good Phase Voice
# We are comparing the average of the different months and if we see a drop then it is likely that the user might churn
total_usage_6 = churn_data[['total_og_mou_6', 'total_ic_mou_6']].sum(axis=1)
total_usage_7 = churn_data[['total_og_mou_7', 'total_ic_mou_7']].sum(axis=1)
total_usage_8 = churn_data[['total_og_mou_8', 'total_ic_mou_8']].sum(axis=1)
mean_6_7 = (total_usage_6+total_usage_7)/2
mean_7_8 = (total_usage_7+total_usage_8)/2
delta_voice = mean_7_8 < mean_6_7
```

```python
churn_data['voice_decline'] = delta_voice.astype(int)
```

```python
mean_offnet_ratio_6_7 = churn_data[['offnet_call_ratio_6','offnet_call_ratio_7']].mean(axis=1)
mean_offnet_ratio_7_8 = churn_data[['offnet_call_ratio_7','offnet_call_ratio_8']].mean(axis=1)
delta_offnet = mean_offnet_ratio_7_8 > mean_offnet_ratio_6_7
churn_data['offnet_ratio_increase'] = delta_voice.astype(int)
churn_data[['offnet_ratio_increase', 'churn']].groupby(['churn']).count()
```

```python
from sklearn.model_selection import train_test_split

# Putting feature variable to X
X = churn_data.drop('churn',axis=1)

# Putting response variable to y
y = churn_data['churn']

y.head()
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
```

```python
#Since there is a high imbalance between non-churners and churners classes, we need to do class-balancing to balance 
#out the records from the two classes. We will use Synthetic Minority Oversampling TEchnique (SMOTE)

from imblearn.over_sampling import SMOTE

print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_sample(X_train, y_train.ravel())

print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
#Improting the PCA module
from sklearn.decomposition import PCA
pca = PCA(svd_solver='randomized', random_state=42)
```

```python
#Doing the PCA on the train data
pca.fit(X_train_res)
```

## Analysis:
### As can be seen from above, only variables related to the action month (month 8) or a few derived metrics were of importance. This indicates that no significant behavior occurred in the good months whatsoever that could have indicated churn. Variables of particular importance from month 8 include those related to recharge amounts for data or calls indicating that significant drops are occurring in these variables, conveying that these major behavioral changes are likely to affect churn outcomes. Also, roaming outgoing calls for month 8 shows up a few times in the decision tree whereas others only show up once. This indicates that it has a higher importance in decision making. What this implies is that people who are more likely to churn are those that need to travel and thus might be facing higher costs compared to other providers. Discounts could be offered to these individuals. We also see some derived metrics with significance. We see that voice_drop has played a significant role which explains the ratio of voice call usage observed in month 8 to the average voice call usage of months 6 and 7. As seen in the bivariate analysis, those who are likely to churn dropped by almost half of their usage in the good period in month 8. Another important derived metric was voice_decline which compares the moving average across the months, if a drop in this average occurs then the customer is likely to churn. One other derived metric was offnet_ratio_increase. If a customer tends to have friends or family who do not use the same provider as them, they are likely to churn if rates to contact these customers are more expensive. Thus, it is advisable to make these rates lower.
